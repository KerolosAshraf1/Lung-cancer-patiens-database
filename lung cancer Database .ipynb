```python
def all_patient(filename):

    

    df = pd.read_csv(filename)

    collection = db.get_collection("all_patient")

    

    

    for x in range(len(df)):

        patient = {'_id': int(df["_id"][x]),

                   'name': df["Name"][x],

                   'surname': df["Surname"][x],

                   'age': int(df["Age"][x]),
                   
                 'smoke': int(df["Smokes"][x]),
                   
                'alchol': int(df["Alkhol"][x]),
            
                'areaQ': int(df["AreaQ"][x]),

                'result': int(df["Result"][x]),


                   
                  }

        collection.insert_one(patient)



```

```python
def patient(filename):
    df = pd.read_csv(filename)
    
    collection = db.get_collection("patient_new")
    
    for x in range(len(df)):
        patient = {'_id': int(df["_id"][x]),
                   'name': df["Name"][x],
                   'surname': df["Surname"][x],
                   'age': int(df["Age"][x])
                  }

        collection.insert_one(patient)
```

```python
def disp_patient():
    
    collection = db.get_collection("patient_new")
    
    result= collection.find()

    
    for doc in result:
        print(doc)
```

```python
def smoke(filename):
    
    df = pd.read_csv(filename)
    
    collection = db.get_collection("smoke_new")
    
    smokes_values = list(np.unique(df["Smokes"].values))
    
    for x in range(len(smokes_values)):
        smoke = {'_id': int(x+1),
                'smoke':int(smokes_values[x]),
                'PID': list(df[df["Smokes"] == smokes_values[x]]["_id"])
                  }
        collection.insert_one(smoke)
```

```python
def disp_smoke():
    smoke_collection = db.get_collection("smoke_new")

    result = smoke_collection.find()

    
    for doc in result:
        print(doc)
```

```python
def Result(filename):
    
    df = pd.read_csv(filename)
    
    client = MongoClient('localhost', 27017)
    db = client.get_database("Lung_cancer_detection")
    collection = db.get_collection("Result_new")
    
    result_values = list(np.unique(df["Result"].values))
    
    for x in range(len(result_values)):
        result = {'_id': int(x+1),
                'result':int(result_values[x]),
                'PID': list(df[df["Result"] == result_values[x]]["_id"])
                  }
        collection.insert_one(result)
    result= collection.find()
    
    for doc in result:
        print(doc)
```

```python
def Exit():
    root.destroy()
def registration_no():
    file=openpyxl.load_workbook ('Student_data.xlsx') 
    sheet=file.active 
    row =sheet.max_row
    max_row_value=sheet.cell (row=row, column=1).value
    try:
        Registration.set(max_row_value+1)
    except:
        Registration.set("1")
def Clear():

    Name.set('')
    Surname.set('')
    AreaQ.set('')
    Alchol.set('')
    Smoke.set('')
    Age.set('')
    Result.set("Select Result")
    registration_no()
    saveButton.config(state = 'normal')


def Insert():
    R1 = Registration.get()
    N1 = Name.get()
    Sur = Surname.get()
    C1 = Result.get()
    Rel = AreaQ.get()
    S1 = Alchol.get()
    Sm = Smoke.get()
    Ag = Age.get()

    if N1 == "" or C1 == "Select Result" or Rel == "" or S1 == "" or R1 == "" or Sur == "" or Sm == "" or Ag == "":
        messagebox.showerror("Error", "Few Data is missing!")
    else:
        inserted_client = {
            "_id": int(R1),
            "name": N1,
            "surname": Sur,
            "age": Ag,
            "smoke": Sm,
            "alcohol": S1,
            "areaQ": Rel,
            "result": C1
        }

        try:
            db.all_patient.insert_one(inserted_client)
            messagebox.showinfo("Success", "Data inserted successfully!")
        except DuplicateKeyError:
            messagebox.showerror("Error", "The ID already exists in the database!")

    



def Delete(): 
    R1 = Registration.get()
    N1 = Name.get()
    Sur = Surname.get()
    C1 = Result.get()
    Rel = AreaQ.get()
    S1 = Alchol.get()
    Sm = Smoke.get()
    Ag = Age.get()
    try:
        # Connect to MongoDB
        client = MongoClient("mongodb://localhost:27017")
        db = client["Lung_cancer_detection"]  # Replace with your valid database name
        collection = db["all_patient"]

        # Delete the document
        result = collection.delete_one({"_id": int(R1)})
        

        if result.deleted_count == 1:
            messagebox.showinfo("Success", "Document deleted successfully!")
        else:
            messagebox.showinfo("Info", "Document not found!")
    except Exception as e:
        messagebox.showerror("Error", str(e))
    finally:
        client.close()
```
